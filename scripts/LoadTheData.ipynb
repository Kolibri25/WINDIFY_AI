```python
# Einlesen der BARRA_C2 Datein

# Pfad zum Datenordner
data_folder = r"C:\Users\Xenia\OneDrive\Documents\2021-2025_BA_Geophysik_Meteorologie_Koeln\2025 SoSe\METFUT_II\Play_around\data\BARRA_C2"

# Monate: Okt, Nov, Dez, Jan
months = ["10", "11", "12", "01"]
years = ["2000", "2001", "2002", "2003"]

for year in years:
    for month in months:
        # Januar beginnt erst ab 2001
        if month == "01" and year == "2000":
            continue
        # Okt–Dez enden mit 2002
        if month in ["10", "11", "12"] and year == "2003":
            continue

        yyyymm = year + month
        filename = f"ua10_AUST-04_ERA5_historical_hres_BOM_BARRA-C2_v1_day_{yyyymm}-{yyyymm}.nc"
        filepath = os.path.join(data_folder, filename)

        if os.path.exists(filepath):
            var_name = f"BARRA_C2_ua10_{yyyymm}"
            globals()[var_name] = xr.open_dataset(filepath)
            print(f"{var_name} erfolgreich geladen.")
        else:
            print(f"Datei nicht gefunden: {filename}")
```

### ERA5 laden

```python
# Hauptdatenordner
base_folder = r"C:\Users\Xenia\OneDrive\Documents\2021-2025_BA_Geophysik_Meteorologie_Koeln\2025 SoSe\METFUT_II\Play_around\data\ERA5"

# Jahre, die geladen werden sollen
years = ["2000", "2001", "2002", "2003"]

# Dateinamen in jedem Jahresordner
filenames = ["data_stream-oper_stepType-instant.nc", "data_stream-wave_stepType-instant.nc"]

# Zwischenspeicher für Einzeldatensätze
datasets_by_year = {}

# Dateien einlesen und Datasets zuweisen
for year in years:
    year_folder = os.path.join(base_folder, f"ERA5_{year}")
    datasets = []
    for fname in filenames:
        filepath = os.path.join(year_folder, fname)
        if os.path.exists(filepath):
            # Gültigen Variablennamen erstellen
            var_basename = fname.replace(".nc", "").replace("-", "_")
            var_name = f"ERA5_{year}_{var_basename}"
            ds = xr.open_dataset(filepath)
            globals()[var_name] = ds
            datasets.append(ds)
            print(f"{var_name} erfolgreich geladen.")
        else:
            print(f"Datei nicht gefunden: {filepath}")

    # Wenn beide Datensätze vorhanden sind, zusammenführen
    if datasets:
        merged_name = f"ERA5_{year}"
        globals()[merged_name] = xr.merge(datasets)
        print(f"{merged_name} erfolgreich zusammengeführt.")
```

```python
# Hauptdatenordner
base_folder = r"C:\Users\Xenia\OneDrive\Documents\2021-2025_BA_Geophysik_Meteorologie_Koeln\2025 SoSe\METFUT_II\Play_around\data\ERA5"

# Jahre, die geladen werden sollen
years = ["2000", "2001", "2002", "2003"]

# Dateinamen der täglichen Daten
daily_filenames = [
    "2m_temperature_stream-oper_daily-mean.nc",
    "mean_sea_level_pressure_0_daily-mean.nc",
    "mean_wave_direction_1_daily-mean.nc",
    "10m_u_component_of_wind_stream-oper_daily-mean.nc",
    "10m_v_component_of_wind_0_daily-mean.nc"
]

# Zwischenspeicher für Datensätze
datasets_by_year = {}

# Dateien einlesen und Datensätze erstellen
for year in years:
    datasets = []

    # Durch alle täglichen Dateien gehen
    for fname in daily_filenames:
        filepath = os.path.join(base_folder, f"ERA5_{year}_daily", fname)

        # Überprüfen, ob die Datei existiert
        if os.path.exists(filepath):
            # Gültigen Variablennamen erstellen
            var_basename = fname.replace(".nc", "").replace("-", "_")
            var_name = f"ERA5_{year}_daily_{var_basename}"

            # Datei einlesen
            ds = xr.open_dataset(filepath)

            # Dataset als globale Variable speichern
            globals()[var_name] = ds
            datasets.append(ds)

            print(f"{var_name} erfolgreich geladen.")
        else:
            print(f"Datei nicht gefunden: {filepath}")

    # Wenn Datensätze vorhanden sind, zusammenführen
    if datasets:
        merged_name = f"ERA5_{year}_daily"
        merged_ds = xr.merge(datasets)
        globals()[merged_name] = merged_ds
        print(f"{merged_name} erfolgreich zusammengeführt.")
```
